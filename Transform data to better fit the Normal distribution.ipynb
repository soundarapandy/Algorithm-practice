```python
#How to Transform Data to Better Fit The Normal Distribution
```

```python
#Tutorial Overview
```

```python
from numpy.random import seed
from numpy.random import randn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import append 
from numpy import zeros
from numpy.random import rand
from numpy import exp
from scipy.stats import boxcox
```

```python
#sample size of mean 100 and SD 50
#seed the random number generator
seed(1)
```

```python
#generate a univariate data sample
data = 50*randn(50)+100
```

```python
plt.hist(data)
plt.show()
```

```python
#sample size from 50 to 100
```

```python
seed(1)
data = 50*randn(100)+100
plt.hist(data)
plt.show()
```

```python
#Data Resolution
```

```python
seed(1)
data =  randn(100)
#Remove decimal component
data = data.round(0)
plt.hist(data)
plt.show()
```

```python
#Extreme Values
```

```python
seed(1)
data = 5*randn(100)+10
#Add extreme values
data = append(data,zeros(10))
plt.hist(data)
plt.show()
```

```python
#Long tails
```

```python
seed(1)
data = 5*randn(100)+10
tail = 10+(rand(50)*100)
#add long tail
data = append(data,tail)
plt.hist(data)
plt.show()
```

```python
#Use simple threshold,such as value of 25 as a cutoff and remove all observations higher than this threshold
```

```python
seed(1)
```

```python
data = 5 * randn(100)+10
tail = 10 +rand(10)*100
data = append(data,tail)
#trim values
data = [x for x in data if x<25]
plt.hist(data)
plt.show()
```

```python
#Power Transforms
```

```python
seed(1)
data = 5*randn(100)+50
#transform to be exponential
data = exp(data)
plt.hist(data)
plt.show()
```

```python
#Box-cox transform
```

```python
seed(1)
data = 5*randn(100)+100
data = exp(data)
data = boxcox(data,0)
plt.hist(data)
plt.show()
```
